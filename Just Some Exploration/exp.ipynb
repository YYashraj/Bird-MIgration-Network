```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import geopandas as gpd


# Read the world shapefile
world_map = Basemap()

# Create a figure and axis with Basemap projection
fig, ax = plt.subplots(figsize=(10, 10))
ax = plt.gca()

# Draw coastlines, countries, and continents
world_map.drawcoastlines()
world_map.drawcountries()
world_map.fillcontinents(color='lightgrey', lake_color='lightblue')

# Plot the data from the shapefile
# Replace 'path/to/data.shp' with the actual path to your shapefile
data = gpd.read_file('data/IBAsGlobal_2023_September_POL_02.shp')
data.plot(ax=ax, color='red', markersize=5)

# Add title
plt.title('Data Over World Map')

# Show the plot
plt.show()
```

```python
import pandas as pd
df2=pd.read_csv('Movebank/filtered/Common-Crane-2020.csv')
df2.columns
```

```python

from shapely.geometry import Point

# Assuming your DataFrame is named df2 and contains 'location-long' and 'location-lat' columns
# Create a GeoDataFrame from the DataFrame
geometry = [Point(xy) for xy in zip(df2['location-long'], df2['location-lat'])]
gdf = gpd.GeoDataFrame(df2, geometry=geometry, crs='EPSG:4326')

# Load the world dataset from geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the world map
world.plot(color='lightgrey', edgecolor='black', figsize=(10, 10))

# Plot the GeoDataFrame containing longitude and latitude coordinates
gdf.plot(ax=plt.gca(), marker='o', color='red', markersize=5, alpha=0.5)
data.plot(ax=plt.gca(), color='blue', markersize=20)


plt.title('Points of Common Crane Sightings in 2020')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(0,50)
plt.ylim(45,70)
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Colors for markers
colors = ['red', 'green', 'olive', 'purple', 'gold', 'salmon', 'turquoise', 'coral', 'lime', 'indigo']

# Create GeoDataFrames
gdfs = []
for i, df in enumerate([df8, df7, df3, group_18, df5, df4, df2]):
    geometry = [Point(xy) for xy in zip(df['location-long'], df['location-lat'])]
    gdf = gpd.GeoDataFrame(df, geometry=geometry, crs='EPSG:4326')
    gdfs.append(gdf)

# Plot world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.plot(color='lightgrey', edgecolor='black', figsize=(10, 10))

# Plot GeoDataFrames
for i, gdf in enumerate(gdfs):
    # Plot as line plot for group_18, df8, and df5
    if 'group_18' in gdf.columns or 'df8' in gdf.columns or 'df5' in gdf.columns:
        gdf.plot(ax=plt.gca(), color=colors[i], linestyle='-', linewidth=1)
    else:
        gdf.plot(ax=plt.gca(), marker='o', color=colors[i], markersize=5, alpha=0.3)

# Plot data DataFrame with reduced transparency
data.plot(ax=plt.gca(), color='blue', markersize=20, alpha=0.2)

plt.legend(['Arctic Shoebird', 'Spotted Eagle', 'Common Crane', 'Swainson\'s Hawk 18', 'Siberian Crane', 'Honey Buzzard', 'Common Crane 2020', 'Common Crane 2016', 'IBA'], loc='lower right', fontsize='small')
plt.title('Points on World Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(-150,150)
plt.ylim(-50,80)
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString

# Colors for markers
colors = ['red', 'green', 'yellow', 'purple', 'gold', 'salmon', 'turquoise', 'coral', 'lime', 'indigo']

# Create GeoDataFrames
gdfs = []
for i, df in enumerate([df8, df7, df3, group_18, df5, df6, df4, df2]):
    geometry = [Point(xy) for xy in zip(df['location-long'], df['location-lat'])]
    # Create LineString connecting consecutive points
    lines = [LineString([(p1.x, p1.y), (p2.x, p2.y)]) for p1, p2 in zip(geometry[:-1], geometry[1:])]
    # Create GeoDataFrame for the lines
    line_gdf = gpd.GeoDataFrame(geometry=lines, crs='EPSG:4326')
    # Set the same color for both markers and edges
    line_gdf['color'] = colors[i]
    gdfs.append((gpd.GeoDataFrame(df, geometry=geometry, crs='EPSG:4326'), line_gdf))

# Plot world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.plot(color='lightgrey', edgecolor='black', figsize=(10, 10))

# Plot GeoDataFrames
for gdf, line_gdf in gdfs:
    # Plot markers
    gdf.plot(ax=plt.gca(), marker='o', color=line_gdf['color'], markersize=5, alpha=0.5)
    # Plot edges
    line_gdf.plot(ax=plt.gca(), color=line_gdf['color'], linestyle='-', linewidth=1, alpha=0.5)

# Plot data DataFrame with reduced transparency
data.plot(ax=plt.gca(), color='blue', markersize=20, alpha=0.2)

plt.title('Points on World Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(-150,150)
plt.ylim(-50,80)
plt.show()
```
